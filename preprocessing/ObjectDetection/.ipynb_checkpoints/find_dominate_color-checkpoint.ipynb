```python
def crop_image(file_name, left_top, right_bottom):
    
    img = Image.open(file_name)
    width, height = img.size
    left = left_top[0] * width
    top = left_top[1] * height
    right = right_bottom[0] * width 
    bottom = right_bottom[1] * height

    crop_img = img.crop((left, top, right, bottom)) 

    plt.imshow(crop_img)
    return crop_img
```
